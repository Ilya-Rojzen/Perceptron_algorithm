```python
import numpy as np
import random
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
# Our dataset is sentences that contain happy and sad words. The number of dominant words in a sentence determines whether 
# it belongs to the 'happy' or 'sad' class. 

features_list = []
labels_list = []

# There will be 100 observations in our dataset
for i in range(100):
    # Number of each feature will be determined randomly
    happy_feat = random.randint(0, 20)
    sad_feat = random.randint(0, 20)

    # Sentence belongs to 'happy' class if number of happy words >= number of sad words
    features = [happy_feat, sad_feat]
    label = 0 if sad_feat > happy_feat else 1
    features_list.append(features)
    labels_list.append(label)

# Generate arrays to further actions
features = np.array(features_list)
labels = np.array(labels_list)

# Generate DataFrame to visualize
df = pd.DataFrame(data=features, columns=['Happy words','Sad words'])
df['Class'] = labels

df.head(5)
```

```python
# Create plot for distribution of classes visualization 

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x='Happy words',
    y='Sad words',
    hue='Class',    
    style='Class',                # Different colors and markers for each class
    palette={0: 'blue', 1: 'red'},  
    markers={0: 'o', 1: 's'},  
    s=100  
)

plt.title('Seaborn plot for vizualization')
plt.grid(True)
plt.show()
```

```python
# Perceptron method 

# Function that returns score for each dot
def score(weights, bias, features):
    return np.dot(features, weights) + bias   # Computes dot product of weights,features and adds bias

# Step function for each dot 
def step(x):
    if x >= 0:
        return 1 
    else:
        return 0

# Function that returns class prediction for each dot
def prediction(weights, bias, features):
    return step(score(weights, bias, features)) 

# Error function that returns 0 for correct prediction and score absolute value for incorrect prediction  
def error(weights, bias, features, label):  
    pred = prediction(weights, bias, features)
    if pred == label:
        return 0 
    else:
        return np.abs(score(weights, bias, features))

# Function that computes error for each dot and returns mean error 
def mean_perceptron_error(weights, bias, features, label):
    total_error = 0
    for i in range(len(features)):
        total_error += error(weights, bias, features[i], label[i])
    return total_error/len(features)

# Function that modulates weights and bias 
def perceptron_method(weights, bias, features, label, learning_rate=0.001):
    pred = prediction(weights, bias, features)
    for i in range(len(weights)):
        weights[i] += (label - pred)*features[i]*learning_rate   
    bias += (label-pred)*learning_rate
    return weights, bias
```

```python
# Perceptron algorithm and visualization

def perceptron_algorithm(features, labels, learning_rate=0.001, epochs=500):
    weights = [1.0 for i in range(len(features[0]))]  # Initializing weights as 1.0
    bias = 0.0                                        # Initializing bias as 0.0
    errors = []
    history = [] # List for further visualization
    for epoch in range(epochs):        
        errors.append(mean_perceptron_error(weights, bias, features, labels))    # Compute MPE and save it
        i = random.randint(0, len(features)-1)   # Choose random dot
        weights, bias = perceptron_method(weights, bias, features[i], labels[i]) # Corrects weights and bias for dot
        history.append((weights.copy(), bias))
    return weights, bias, history, errors

# Function for classifiers training visualization
def plot_decision_boundaries(features, labels, history):
    plt.figure(figsize=(10, 6))
    # Our dots
    plt.scatter(features[:, 0], features[:, 1], c=labels, cmap='bwr', edgecolors='k')
    
    # Borders for visualization
    x_min, x_max = features[:, 0].min() - 1, features[:, 0].max() + 1
    y_min, y_max = features[:, 1].min() - 1, features[:, 1].max() + 1
    
    for i, (w, b) in enumerate(history):
        
        # Line equation: w0*x + w1*y + b = 0 → y = (-w0/w1)x - (b/w1)
        x_values = np.array([x_min, x_max])
        y_values = (-w[0]/w[1])*x_values - b/w[1]
        
        plt.plot(x_values, y_values, 'g:', alpha=i/len(history), linewidth=0.5)

    # Final classifier
    final_w, final_b = history[-1]
    x_values = np.array([x_min, x_max])
    y_values = (-final_w[0]/final_w[1])*x_values - final_b/final_w[1]
    plt.plot(x_values, y_values, 'r-', linewidth=2, label='Final boundary')

    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Evolution of perceptron classifiers')
    plt.legend()
    plt.grid(True)
    plt.show()
    plt.savefig('./pictures/Perceptron_evolution.png')
    
# Function for MPE visualization
def plot_loss_history(loss_history):
    plt.figure(figsize=(10, 6))
    plt.plot(loss_history, 'r-', linewidth=2)
    plt.xlabel('Epoch')
    plt.ylabel('MPE')
    plt.title('MPE evolution')
    plt.grid(True)
    plt.show()
    plt.savefig('./pictures/MPE_evolution.png')
```

```python
history = perceptron_algorithm(features, labels)[-2]
plot_decision_boundaries(features, labels, history=history)
```

```python
errors = perceptron_algorithm(features, labels)[-1]
plot_loss_history(errors)
```
